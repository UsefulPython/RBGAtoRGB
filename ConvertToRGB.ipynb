```python
from PIL import Image
```

```python
def convertRGBAToRGB(rgbaImage, name):
    # Print the image name 
    print(rgbaImage)
    # Open the image
    rImage = Image.open(rgbaImage)
    # Display the image Type size and mode
    print("Original Image:",rImage.format,rImage.size, rImage.mode)
    
    # load a blank image wi
    background = Image.new("RGB", rImage.size, (255,255,255))
    background.paste(rImage, mask = rImage.split()[3])
    saveName = name + rgbaImage
    background.save(saveName, "JPEG", quality = 100)
    rbgImage = Image.open(saveName)
    rbgImage
    print("Converted Image:", rbgImage.format,rbgImage.size, rbgImage.mode)
    return rbgImage
```

```python
convertedImage = convertRGBAToRGB("raspberry.png", "raspberrySuccess")
convertedImage
```
